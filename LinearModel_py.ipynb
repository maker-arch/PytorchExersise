```python
import numpy as np
import matplotlib.pyplot as plt

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

def forward(x):
  return x * w

def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) * (y_pred - y)

w_list = []
loss_list = []

for w in np.arange(0.1, 4.1, 0.1): # w是全局变量
  print('w=', w)
  l_sum = 0
  for x_val, y_val in zip(x_data, y_data):
    y_pred_val = forward(x_val)
    loss_val = loss(x_val, y_val)
    l_sum += loss_val
    print('\t', x_val, y_val, y_pred_val, loss_val)
  print('mse_loss=', l_sum/3)
  w_list.append(w)
  loss_list.append(l_sum/3)

plt.plot(w_list, loss_list)
plt.xlabel('w')
plt.ylabel("loss")
plt.show()
```

```python

x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

def forward(x):
  return x * W + B

def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) * (y_pred - y)

loss_list = []
w_list = np.arange(0.0, 4.1, 0.1) #左闭又开
b_list = np.arange(-2.0, 2.1, 0.5)
W, B = np.meshgrid(w_list, b_list)


l_sum = 0
for x_val, y_val in zip(x_data, y_data):
  y_pred_val = forward(x_val)
  loss_val = loss(x_val, y_val)
  l_sum += loss_val

fig = plt.figure(figsize=(16, 16))
ax = fig.add_subplot(projection='3d')
ax.plot_surface(W,B,l_sum/3, cmap='viridis')
ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('Loss', fontsize=5)
plt.title('Loss Surface')
plt.show()
```

```python
import numpy as np

x = np.array([1,2,3])   # Nx=3
y = np.array([4,5])     # Ny=2

# 1. 默认：indexing='xy', sparse=False
X1, Y1 = np.meshgrid(x, y)
print("=== indexing='xy' 完整网格 ===")
print("X1:\n", X1)
print("Y1:\n", Y1)
print("形状 X1", X1.shape)   # (2,3)

# 2. indexing='ij'
X2, Y2 = np.meshgrid(x, y, indexing='ij')
print("\n=== indexing='ij' ===")
print("X2:\n", X2)
print("Y2:\n", Y2)
print("X2.shape", X2.shape)   # (3,2)

# 3. sparse=True 稀疏模式
X3, Y3 = np.meshgrid(x, y, sparse=True)
print("\n=== sparse=True 稀疏 ===")
print("X3:\n", X3)
print("Y3:\n", Y3)
print("X3.shape", X3.shape)
print("Y3.shape", Y3.shape)
```
