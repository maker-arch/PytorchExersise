```python
import torch

#定义模型
class LinearModel(torch.nn.Module):
  def __init__(self):
    super(LinearModel, self).__init__()
    self.linear = torch.nn.Linear(1,1)

  def forward(self, x):
    y_pred = self.linear(x)
    return y_pred

model = LinearModel()
#定义损失和优化器
criterion = torch.nn.MSELoss(size_average=False)
optimizer = torch.optim.SGD(model.parameters(), lr=0.03)

#加载数据集
x_data = torch.Tensor([[1.0],[2.0],[3.0]])
y_data = torch.Tensor([[2.0],[4.0],[6.0]])

#训练
for epoch in range(100):
  y_pred = (model(x_data))
  loss = criterion(y_pred, y_data)
  print(epoch,loss.item())

  optimizer.zero_grad()
  loss.backward()
  optimizer.step()
print("w=", model.linear.weight.item())
print("b=", model.linear.weight.item())

x_test = torch.Tensor([[4.0]])
y_test = model.forward(x_test)

print("y_pred:", y_test.data)
```

```python
import torch
import matplotlib.pyplot as plt

class LinearModel(torch.nn.Module):
  def __init__(self):
    super(LinearModel, self).__init__()
    self.linear = torch.nn.Linear(1,1)

  def forward(self, x):
    y_pred = self.linear(x)
    return y_pred
x_data = torch.Tensor([[1.0], [2.0], [3.0]])
y_data = torch.Tensor([[2.0, 3.0, 6.0]])

optimizers = {
    "SGD": torch.optim.SGD,
    "Adagrad": torch.optim.Adagrad,
    "Adam": torch.optim.Adam,
    "Adamax": torch.optim.Adamax,
    "ASGD": torch.optim.ASGD,
    "LBFGS": torch.optim.LBFGS,
    "RMSprop": torch.optim.RMSprop,
    "Rprop": torch.optim.Rprop
}

epochs = 100
loss_history = {}

for opt_name, opt_class in optimizers.items():
  model = LinearModel()
  criterion = torch.nn.MSELoss()

  optimizer = opt_class(model.parameters(), lr=0.01)

  losses = []
  print(f"\n====== 训练优化器：{opt_name} ======")

  for epoch in range(epochs):
    y_pred = model(x_data)
    loss = criterion(y_pred, y_data)
    losses.append(loss.detach().item()) #detach() 剥离梯度，防止numpy报错

    optimizer.zero_grad()

    if opt_name=="LBFGS":
      def closure():
        optimizer.zero_grad()
        y_pred = model(x_data)
        loss = criterion(y_pred, y_data)
        loss.backward()
        optimizer.step()
    else:
      loss.backward()
      optimizer.step()


    if epoch % 10 == 0:
      print(f"epoch: {epoch}, loss: {loss.item()}")

  loss_history[opt_name] = losses


plt.figure(figsize=(12, 7))

colors = ['b', 'g', 'r', 'c', 'm', 'y', 'k', 'orange']

for i, (name, losses) in enumerate(loss_history.items()):
  plt.plot(range(epochs), losses, label=name, color=colors[i], linewidth=2)

plt.xlabel("epoch", fontsize=12)
plt.ylabel("loss(MSE)", fontsize=12)
plt.title("loss", fontsize=14)

plt.legend(fontsize=10)
plt.grid(True, alpha=0.3)
plt.show()




```
