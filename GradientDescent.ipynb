```python
import numpy as np
```

```python
# 全部的gradient descent
x_data = [1.0, 2.0, 3.0]
y_data = [2.0, 4.0, 6.0]

# 初始化参数
w = 1.0
#前馈函数
def forward(x):
  return x * w

# 注意下全局变量
def cost(xs, ys): #均值
  cost = 0
  for x_val, y_val in zip(xs, ys):
    y_pred = forward(x_val)
    cost += (y_pred - y_val) ** 2
  return cost / len(xs)

def gradient(xs, ys):
  grad = 0
  for x_val, y_val in zip(xs, ys):
    grad += 2 * x_val * (x_val * w - y_val)

  return grad / len(xs)


print('before training predict of x=4 is ', forward(4))
for epoch in range(100):

  grad_val = gradient(x_data, y_data)

  w -= 0.01 * grad_val
  cost_val = cost(x_data, y_data)
  print('epoch=', epoch, 'w=', w, 'loss=', cost_val)

print('after training predict of x=4 is ', forward(4))



```

```python
# 随机梯度下降 SGD
x_data = [1.0,2.0,3.0]
y_data = [2.0,4.0,6.0]

w = 1.0
def forward(x):
  return x * w

def loss(x, y):
  y_pred = forward(x)
  return (y_pred - y) ** 2

def grad(x,y):
  return 2 * x * (x * w - y)

print('before training predict of 4 is ', forward(4))

for epoch in range(100):
  for x_val, yval in zip(x_data, y_data):
    gradient = grad(x_val, yval)

    w -= 0.01 * gradient

    l = loss(x_val, yval)
  print('epoch=', epoch, 'w=', w, 'loss=',l)

print('before training predict of 4 is ', forward(4))
```
